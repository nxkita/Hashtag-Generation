### Scraping

#### Selenium is used to create a temporary browser that scrapes images from Instagram's latest posts, based on the number specified. It navigates the hashtag's pages, grabs the links, images and hashtags through the defined functions and saves it to 'data' and 'metadata' folders.

```python
hashtags = ["travel", "food", "animals", "selfie", "cars", "coffee", "babies", "wedding", "nature", "architecture"]

num_to_scrape = 8
```

```python
folder_names = ["data", "metadata"]
for folder_name in folder_names:
    try:
        os.mkdir(folder_name)
    except OSError:
        print(f"Folder '{folder_name}' already exists.")
```

```python
scrape_data(hashtags, num_to_scrape, delay=5)
```
